```python
city_station_dict = {}
for i in range(len(station_options)):
    city = station_options[0][i].split("'")[1]
    station_code = station_options[1][i].split("'")[1]
    city_station_dict[city] = station_code
```

```python
for city in cities:
    if city == "Петропавловск":
        print("Петропавловск-Камчатский" in city_station_dict.keys())
    elif city == "Орел":
        print("Орёл" in city_station_dict.keys())
    elif not city in city_station_dict.keys():
        print(city)
```

```python
def parse_weather(code, city):
    url = f"http://pogoda-service.ru/archive_gsod_res.php?country=RS&station={code}&datepicker_beg=01.01.2015&datepicker_end=01.01.2020&bsubmit=%D0%9F%D0%BE%D1%81%D0%BC%D0%BE%D1%82%D1%80%D0%B5%D1%82%D1%8C"
    data = pd.read_html(url)
    df = data[0]
    df.columns = ["Дата", "Максимальная температура", "Минимальная температура", "Средняя температура", "Атмосферное давление", "Скорость ветра", "Осадки", "Эффективная температура"]
    df['Дата'] = pd.to_datetime(df['Дата'], format='%d.%m.%Y')
    df['Город'] = city
    return df
```

```python
df_weather = pd.DataFrame(columns = ["Дата", "Максимальная температура", "Минимальная температура", "Средняя температура", "Атмосферное давление", "Скорость ветра", "Осадки", "Эффективная температура", 'Город'])
for city in tqdm(cities):
    if city == "Петропавловск":
        code = city_station_dict["Петропавловск-Камчатский"]
        df_weather = df_weather.append(parse_weather(code=code, city=city), ignore_index = True)

    elif city == "Орел":
        code = city_station_dict["Орёл"]
        df_weather = df_weather.append(parse_weather(code=code, city=city), ignore_index = True)
        
    elif city in city_station_dict.keys():
        code = city_station_dict[city]
        df_weather = df_weather.append(parse_weather(code=code, city=city), ignore_index = True)

    else:
        print(city)
```

```python
df_weather.to_csv('weather_by_city.csv', index=False, sep=';', encoding='utf-8-sig')
```
